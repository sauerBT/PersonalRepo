Exercise 4: Data Selection and the .loc Method

Import numpy and the pandas library using the following command in the Jupyter Notebook:

```python
import numpy as np
import pandas as pd
```

Read the RadNet dataset from the EPA, available from the Socrata project:

```python
url = "radData.csv"
df = pd.read_csv(url)
```

Select a column

```python
df["Location"].head()
```

Filter the selected values in the "State" column using the "MN" value 

```python
df[df["State"] == "MN"].head()
```

Use multiple conditions to filter the dataset

```python
df[(df["State"] == "MN") & (df["Location"] == "St. Paul")].head()
```

Select the "MN" states and isotope I-131 column

```python
df[df["State"] == "MN"]["I-131"].head()
```

Do the selection using the .loc method.  The difference is that the result of the .loc method is a series, not a dataframe

```python
df.loc[df["State"] == "MN", "I-131"].head()
```
